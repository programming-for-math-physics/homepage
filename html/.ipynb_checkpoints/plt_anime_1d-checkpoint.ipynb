```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import math,time

# 復習: アニメなしで y = sin(x)を表示
def static_1d():
    x = np.linspace(0.0, 2.0 * math.pi, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.show()

# アニメあり y = sin(k * x) を k を色々変えながら表示
def anime_1d():
    x = np.linspace(0.0, 2.0 * math.pi, 100)
    y = np.sin(x)
    # 一度だけプロットして曲線を得る.
    # plotの結果(線)を変数で受け取る．
    # 実はplotはリストを返すので，左辺は[line] = 
    # とする
    [line] = plt.plot(x, y)
    for k in range(10):
        # アニメーションするときは，plotを呼ぶ
        # plotする代わりにデータをセット(set_data)する
        line.set_data(x, np.sin(k * x))
        # 再描画して欲しいタイミングで， yieldを呼ぶ
        # yield には，変更された曲線のリストを渡す
        yield [line]

def anime_1dx():
    x = np.linspace(0.0, 2.0 * math.pi, 100)
    fig = plt.figure()
    plots = []
    for k in range(10):
        # アニメーションするときは，plotを呼ぶ
        # plotする代わりにデータをセット(set_data)する
        plots.append(plt.plot(x, np.sin(k * x)))
    ani = animation.ArtistAnimation(fig, plots, interval=100)
    plt.show()
    return ani

# おまじない(コピペして下さい．これまでの範囲では
# 理解不能ですので悩む必要なし)
def go_animation(iterator, **kwargs):
    def anime_fun(*args):
        try:
            return next(iterator)
        except StopIteration:
            return []
    ani = animation.FuncAnimation(plt.gcf(), anime_fun, **kwargs)
    plt.show()

# sin(kx) のグラフが徐々に変わっていく
#go_animation(anime_1d(), interval=100)
anime_1dx()

```
